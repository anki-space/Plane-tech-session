**1. Tokenization**

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize



# Your text
text = "AI is revolutionizing the world of technology! NLP is a key part of AI."

# Word Tokenization
word_tokens = nltk.tokenize.word_tokenize(text)
print("Word Tokens:", word_tokens)

# Sentence Tokenization
sentence_tokens = sent_tokenize(text)
print("Sentence Tokens:", sentence_tokens)
```

**2. Stopwords**

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))  # Load English stopwords

# Example text
tokens = word_tokenize("This is an amazing NLP tutorial for beginners!")

# Removing stopwords
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]#list  comprehesion

print("Original Tokens:", tokens)
print("Filtered Tokens (Stopwords removed):", filtered_tokens)
```

**3. Stemming**

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = ["running", "studies", "beautiful", "flies", "coying","crying"]
#y is removed because its previos character is a consonant
stemmed_words = [stemmer.stem(word) for word in words]

print("Stemmed Words:", stemmed_words)
```

**4. Lemmatization**

```python
import spacy
nlp = spacy.load("en_core_web_sm")

text = "Elon Musk founded SpaceX in 2002 and Tesla in 2003."
doc = nlp(text)


for ent in doc.ents:
    print(ent.text, "-", ent.label_)
```

**Sentiment Analysis**

```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
sia = SentimentIntensityAnalyzer()
text = "le bhadwe : I ordered the laptop but the delivery was of a pressure cooker. Totally disappointed."
sentiment = sia.polarity_scores(text)

print("Sentiment Scores:", sentiment)
```

**Simple Chatbot**

```python
from nltk.chat.util import Chat, reflections

pairs = [
    ["hi|hello", ["Hello!", "Hi there!"]],
    ["how are you?", ["I'm a bot, I'm always good!"]],
    ["what's your name?", ["I'm a simple chatbot!"]],
    ["bye", ["Goodbye!", "See you later!"]],
    ["default", ["I'm sorry, I didn't understand you."]],
]
chatbot = Chat(pairs, reflections)
chatbot.converse()
```
